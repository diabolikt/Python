```python
from netaddr import *
import pprint
network202disp = []
network203disp = []
network202 = []
network203 = []
i=j=1
ip202 = IPNetwork('10.202.0.0')
ip203 = IPNetwork('10.203.0.0')
for ip202 in IPNetwork('10.202.0.0/25'):
    iptemp = str(ip202)
    if iptemp[-2:] != '.0' and i<=20:
        network202.append(str(iptemp))
        i=i+1
    elif iptemp[-2:] != '.0':
        network202disp.append(str(iptemp))
        i=i+1
for ip203 in IPNetwork('10.203.0.0/25'):
    iptemp = str(ip203)
    if iptemp[-2:] != '.0' and j<=20:
        network203.append(str(iptemp))
        j=j+1
    elif iptemp[-2:] != '.0':
        network203disp.append(str(iptemp))
print('IP utilise 202:',network202)
print ('')
print('IP utilise 203:',network203)
print ('')
print('IP dispo 202:',network202disp)
print ('')
print('IP dispo 203:',network203disp)
```
